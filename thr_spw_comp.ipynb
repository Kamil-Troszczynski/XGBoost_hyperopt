```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_diff_thresh_02 = pd.read_csv('metrics/gbtree/Optuna/thresh_spw_archive/metrics_2026-01-17 19:16:19.877252.csv')
df_diff_thresh_05 = pd.read_csv('metrics/gbtree/Optuna/metrics_2026-01-17 21:58:02.446496.csv')
df_diff_thresh_08 = pd.read_csv('metrics/gbtree/Optuna/thresh_spw_archive/metrics_2026-01-17 18:58:23.729863.csv')

df_diff_spw__4 = pd.read_csv('metrics/gbtree/Optuna/thresh_spw_archive/metrics_2026-01-17 19:31:42.999216.csv')
df_diff_spw    = pd.read_csv('metrics/gbtree/Optuna/metrics_2026-01-17 21:58:02.446496.csv')
df_diff_spw_4  = pd.read_csv('metrics/gbtree/Optuna/thresh_spw_archive/metrics_2026-01-17 19:49:57.263060.csv')
```

```python
time = np.arange(0, df_diff_thresh_02.shape[0], 1)


#   DIFFERENT THRESHOLD
acc1, acc2, acc3 = df_diff_thresh_02['accuracy'], df_diff_thresh_05['accuracy'], df_diff_thresh_08['accuracy']
prec1,prec2,prec3 = df_diff_thresh_02['precision'], df_diff_thresh_05['precision'], df_diff_thresh_08['precision']
rec1, rec2, rec3 = df_diff_thresh_02['recall'], df_diff_thresh_05['recall'], df_diff_thresh_08['recall']
f11, f12, f13 = df_diff_thresh_02['f1'], df_diff_thresh_05['f1'], df_diff_thresh_08['f1']
lc1, lc2, lc3 = df_diff_thresh_02['learning curve'], df_diff_thresh_05['learning curve'], df_diff_thresh_08['learning curve']
gini1, gini2, gini3 = df_diff_thresh_02['gini'], df_diff_thresh_05['gini'], df_diff_thresh_08['gini']


metrics = [
    ('Accuracy', acc1, acc2, acc3),
    ('Precision', prec1, prec2, prec3),
    ('Recall', rec1, rec2, rec3),
    ('F1', f11, f12, f13),
    ('Learning Curve', lc1, lc2, lc3),
    ('Normalized Gini Coeff', gini1, gini2, gini3)
           ]
```

```python
for metric_tuple in metrics:
    metric_name, metric_vector1, metric_vector2, metric_vector3 = (metric_tuple[0], 
                                                                metric_tuple[1], 
                                                                metric_tuple[2], 
                                                                metric_tuple[3])
    plt.figure()
    plt.scatter(time, metric_vector1, label = 'THRESHOLD = 0.2')
    plt.scatter(time, metric_vector2, label = 'THRESHOLD = 0.5')
    plt.scatter(time, metric_vector3, label = 'THRESHOLD = 0.8')
    plt.title(f'{metric_name} for different threshold value')
    plt.ylabel(metric_name)
    plt.xlabel('Trial')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
for metric_tuple in metrics:
    metric_name, metric_vector1, metric_vector2, metric_vector3 = (metric_tuple[0], 
                                                                metric_tuple[1], 
                                                                metric_tuple[2], 
                                                                metric_tuple[3])
    plt.figure()
    plt.scatter(time, metric_vector1, label = 'scale_pos_weight = 22.4')
    plt.scatter(time, metric_vector2, label = 'scale_pos_weight = 26.4') #scale_pos_weight = (sum(y_train == 0) / sum(y_train == 1)) 
    plt.scatter(time, metric_vector3, label = 'scale_pos_weight = 30.4')
    plt.title(f'{metric_name} for different scale_pos_weight value')
    plt.ylabel(metric_name)
    plt.xlabel('Trial')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```
